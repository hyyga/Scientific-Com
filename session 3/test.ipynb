```python
import numpy as np
import matplotlib.pyplot as plt

# Data input
x = np.array([
    0.667257363, 1.163272104, 2.059775843, 3.026043396, 3.823459252, 4.237522149,
    5.175051771, 5.274506600, 6.244287409, 6.542814643, 7.055059567, 7.390116752,
    8.288971198, 8.579231789, 9.210386935, 9.620823117, 10.25700941, 10.89194207,
    11.21369076, 11.41558731, 11.84950537, 12.41654640, 13.07341967, 13.93756043,
    13.96383483, 14.49473702, 14.90972704, 15.07248603, 15.79685930, 16.65678362,
    17.30883901, 17.35006975, 17.58938929, 17.79778379, 18.00169342, 18.48060630,
    18.71248139, 19.11376774, 19.39409803, 19.74650750, 20.74546216, 21.01455518,
    21.69457689, 22.33519718, 23.26208851, 23.97298783, 24.60252331, 25.28542718,
    25.36556687, 25.43342769
])
y = np.array([
    69.61612841, 70.31009462, 70.52967130, 70.69226554, 71.26272117, 72.19410464,
    72.75925445, 73.48095621, 74.32835598, 75.00604086, 75.81053537, 76.18342777,
    76.26701376, 76.88125364, 76.88850065, 77.20628611, 77.99029850, 78.41087441,
    79.33575636, 79.98614235, 80.62260103, 81.46225811, 81.97599388, 82.40526228,
    82.75687635, 83.19695143, 83.22854080, 84.05539698, 84.18519963, 84.20141506,
    84.96388936, 85.96117288, 86.20746789, 86.59131925, 87.14506062, 87.80712240,
    88.51922370, 88.82890416, 89.52846481, 90.37898232, 90.98286205, 91.47855750,
    91.71240983, 91.87073017, 91.94818226, 92.59044238, 93.44106750, 93.50315021,
    94.20888309, 94.39807859
])

# Membentuk matriks desain A
A = np.vstack([x, np.ones(len(x))]).T  # Menumpuk x dan konstanta 1, kemudian ditranspose

# Menambahkan dimensi baru pada y (jika diperlukan)
y = y[:, np.newaxis]

# Perhitungan regresi menggunakan (A^T A)^-1 A^T y
a1 = np.dot(A.T, A)           # A^T A
a2 = np.linalg.inv(a1)        # (A^T A)^-1
a3 = np.dot(a2, A.T)          # (A^T A)^-1 A^T
a4 = np.dot(a3, y)            # (A^T A)^-1 A^T y

# Hasil parameter regresi
slope, intercept = a4.flatten()  # Ekstrak slope dan intercept
print(f"Slope (m): {slope}, Intercept (c): {intercept}")

# Visualisasi hasil
plt.scatter(x, y, label="Data")
plt.plot(x, slope * x + intercept, color='red', label=f"y = {slope:.2f}x + {intercept:.2f}")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.show()
```
